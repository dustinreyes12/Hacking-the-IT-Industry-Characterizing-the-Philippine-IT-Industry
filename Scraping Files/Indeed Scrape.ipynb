```python
sesh = requests.session()
sesh.headers.update(headers)

job_url = []
for pg in range(0,991,10):
    jobs = sesh.get(f'https://www.indeed.com.ph/jobs?q=it&start={pg}')
    jobs_sp = BeautifulSoup(jobs.text,'lxml')
    job_ads = jobs_sp.findAll('div', {'class':['jobsearch-SerpJobCard', 'unifiedRow', 'row', 'result']})
    sleep(5)
    for item in job_ads:
        titleurl = item.findAll('div',{'class':'title'})[0]
        job_url.append(r'https://www.indeed.com.ph'+
                       titleurl.findChildren('a')[0]['href'])
    print(len(job_url))
```

```python
sesh = requests.session()
sesh.headers.update(headers)

job_date = []
for pg in range(0,991,10):
    jobs = sesh.get(f'https://www.indeed.com.ph/jobs?q=it&jt=fulltime&sort=date&start={pg}')
    jobs_sp = BeautifulSoup(jobs.text,'lxml')
    job_ads = jobs_sp.findAll('div',
                          {'class':['jobsearch-SerpJobCard',
                                    'unifiedRow', 'row', 'result']})
    sleep(5)
    for item in job_ads:
        titleurl = item.findAll('div',{'class':'title'})[0]
        job_date.append(r'https://www.indeed.com.ph'+
                       titleurl.findChildren('a')[0]['href'])
    print(len(job_date))
```

```python
job_total = job_url + job_date
```
